```python
import zipfile
import os

zip_files = ['/content/Temp_main.zip', '/content/temp_org.zip']

extracted_dirs = ['/content/Temp_main', '/content/temp_org']

# Extract the zip files
for zip_path, extract_dir in zip(zip_files, extracted_dirs):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)
        print(f"Extracted {zip_path} to {extract_dir}")

for directory in extracted_dirs:
    print(f"Contents of {directory}:")
    print(os.listdir(directory))
```

```python
import cv2
import os
from tqdm import tqdm

def video_to_frames(video_path, base_output_dir):
    video_name = os.path.splitext(os.path.basename(video_path))[0].replace(' ', '_')
    output_dir = os.path.join(base_output_dir, video_name)

    # Ensure the output directory for this video exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    video_capture = cv2.VideoCapture(video_path)

    if not video_capture.isOpened():
        print(f"Error: Cannot open video file {video_path}")
        return

    frame_rate = int(video_capture.get(cv2.CAP_PROP_FPS))
    if frame_rate == 0:
        print(f"Warning: Frame rate is zero for {video_path}")
        video_capture.release()
        return

    frame_count = 0
    extracted_count = 0

    while True:
        ret, frame = video_capture.read()
        if not ret:
            break

        # Extract one frame per second of video
        if frame_count % frame_rate == 0:
            output_frame_path = os.path.join(output_dir, f"frame_{extracted_count}.jpg")
            cv2.imwrite(output_frame_path, frame)
            extracted_count += 1

        frame_count += 1

    video_capture.release()
    print(f"Extracted {extracted_count} frames from {video_path}")

video_dirs = ['/content/Temp_main/Temp_main', '/content/temp_org/temp_org']
base_output_dirs = ['/content/Temp_main_frames', '/content/temp_org_frames']

for video_dir, base_output_dir in zip(video_dirs, base_output_dirs):
    if not os.path.exists(video_dir):
        print(f"Video directory does not exist: {video_dir}")
        continue

    video_files = [f for f in os.listdir(video_dir) if f.lower().endswith('.mp4')]

    if not video_files:
        print(f"No .mp4 files found in directory: {video_dir}")
        continue

    for video_path in tqdm(video_files, desc=f"Processing videos in {video_dir}"):
        full_video_path = os.path.join(video_dir, video_path)
        video_to_frames(full_video_path, base_output_dir)

for base_output_dir in base_output_dirs:
    if os.path.exists(base_output_dir):
        print(f"Directory structure in {base_output_dir}:")
        for dir_name in os.listdir(base_output_dir):
            dir_path = os.path.join(base_output_dir, dir_name)
            extracted_files = [f for f in os.listdir(dir_path) if f.endswith('.jpg')]
            print(f"Contents of {dir_name}: {len(extracted_files)} frames extracted.")
            if extracted_files:
                print('\n'.join(extracted_files[:5]))
            else:
                print("No frames were extracted.")
    else:
        print(f"Output directory does not exist: {base_output_dir}")
```

```python
from mtcnn.mtcnn import MTCNN
from PIL import Image
import numpy as np
import cv2
from tqdm import tqdm
import os

def crop_and_resize(input_dir, output_dir, size=(224, 224)):
    # Create the output directory if it does not exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Initialize the MTCNN detector
    detector = MTCNN()

    # Loop through each image in the input directory
    for img_file in tqdm(os.listdir(input_dir)):
        if img_file.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')):
            img_path = os.path.join(input_dir, img_file)
            img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
            img = Image.fromarray(img)

            # Detect faces in the image
            faces = detector.detect_faces(np.array(img))
            if faces:
                x, y, width, height = faces[0]['box']
                face = img.crop((x, y, x + width, y + height))
                face = face.resize(size)
                face.save(os.path.join(output_dir, img_file))

# Base directories for frames and face outputs
base_input_dirs = ['/content/Temp_main_frames', '/content/temp_org_frames']
base_output_dirs = ['/content/Temp_main_faces', '/content/temp_org_faces']

# Process each base directory
for base_input_dir, base_output_dir in zip(base_input_dirs, base_output_dirs):
    if not os.path.exists(base_input_dir):
        print(f"Input directory does not exist: {base_input_dir}")
        continue

    # Process each subdirectory corresponding to a video
    for video_name in os.listdir(base_input_dir):
        input_dir = os.path.join(base_input_dir, video_name)
        output_dir = os.path.join(base_output_dir, video_name)

        if not os.path.exists(input_dir):
            print(f"Video input directory does not exist: {input_dir}")
            continue

        print(f"Processing {input_dir}...")
        crop_and_resize(input_dir, output_dir)
```

```python
import os
import shutil

root_dir = '/content/Temp_main_faces'

subfolders = [f.name for f in os.scandir(root_dir) if f.is_dir()]


if not subfolders:
    os.makedirs(os.path.join(root_dir, 'all_images'), exist_ok=True)  
    for file in os.listdir(root_dir):
        if file.endswith(('.png', '.jpg', '.jpeg')):
            shutil.move(os.path.join(root_dir, file), os.path.join(root_dir, 'all_images'))


from torchvision.datasets import ImageFolder
from torchvision import transforms
from torch.utils.data import DataLoader

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

test_dataset = ImageFolder(root=root_dir, transform=transform)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

def check_data_loader_dim(loader):
    for images, labels in loader:
        print('Image batch dimensions:', images.shape)
        print('Image label dimensions:', labels.shape)
        break

check_data_loader_dim(test_loader)
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split


def create_split(source_dir, train_dir, test_dir, test_size=0.2):
    # Get the list of all files
    files = [os.path.join(source_dir, f) for f in os.listdir(source_dir)]
    train_files, test_files = train_test_split(files, test_size=test_size)

    # Copy files to the new train/test directories
    for f in train_files:
        shutil.move(f, os.path.join(train_dir, os.path.basename(f)))

    for f in test_files:
        shutil.move(f, os.path.join(test_dir, os.path.basename(f)))

os.makedirs('/content/data/train/real', exist_ok=True)
os.makedirs('/content/data/train/fake', exist_ok=True)
os.makedirs('/content/data/test/real', exist_ok=True)
os.makedirs('/content/data/test/fake', exist_ok=True)

create_split('/content/temp_org_faces', '/content/data/train/real', '/content/data/test/real')
create_split('/content/Temp_main_faces', '/content/data/train/fake', '/content/data/test/fake')
```
